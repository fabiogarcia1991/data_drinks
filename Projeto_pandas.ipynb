```python
pd.DataFrame()
df_alcoolismo = pd.DataFrame(df_doenças_alcool)
```

```python
df_alcoolismo.nlargest(5, 'incidência_doença_hepática')
```

```python
df_alcoolismo.nlargest(5,'porções_destilados_ano')
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib
df_alcoolismo.groupby('país')['incidência_doença_hepática'].sum().sort_values(ascending=False).plot.bar(title='incidência de doença hepática')
plt.xlabel('países')
plt.ylabel('latas de cerveja per capta');
```

```python
df_alcoolismo.groupby('país')['porções_destilados_ano'].sum().sort_values(ascending=False).plot.bar(title='maiores consumidores de bebidas destiladas')
plt.xlabel('países')
plt.ylabel('porções / ano');
```

```python
relação_alcool_cirrose = df_alcoolismo.groupby(['consumo_alcool_percapta_L', 'incidência_doença_hepática'])['país'].sum().reset_index()
relação_alcool_cirrose.head(13)
```

```python
relação_alcool_cirrose.plot.bar(x='país')
```
